```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

FEATURES = ['sorted', 'avltree', 'hashtable']
COLORS = ['red', 'green', 'blue']
```

```python
create_all = pd.read_csv("create_all.csv", index_col=0)
find_one = pd.read_csv("find_one.csv", index_col=0)
find_random = pd.read_csv("find_random.csv", index_col=0)
delete_all = pd.read_csv("delete_all.csv", index_col=0)
delete_by_one = pd.read_csv("delete_by_one.csv", index_col=0)
```

```python
create_all
```

```python
plt.figure(figsize=(15, 6))
for feature, color in zip(FEATURES, COLORS):
    plt.plot(create_all[feature].index, create_all[feature].values, label=feature, color=color)
plt.ylabel("milliseconds")
plt.xlabel("count of objects")
plt.title("create all")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
for feature, color in zip(FEATURES, COLORS):
    plt.plot(find_one[feature].index, find_one[feature].values, label=feature, color=color)
plt.ylabel("milliseconds")
plt.xlabel("count of objects")
plt.title("find one")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
for feature, color in zip(FEATURES, COLORS):
    plt.plot(find_random[feature].index, find_random[feature].values, label=feature, color=color)
plt.ylabel("milliseconds")
plt.xlabel("count of objects")
plt.title("find random")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))
for feature, color in zip(FEATURES, COLORS):
    plt.plot(delete_all[feature].index, delete_all[feature].values, label=feature, color=color)
plt.ylabel("milliseconds")
plt.xlabel("count of objects")
plt.title("delete all")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 6))

for feature, color in zip(FEATURES[1:], COLORS[1:]):
    plt.plot(delete_by_one[feature].index, delete_by_one[feature].values, label=feature, color=color)
plt.ylabel("milliseconds")
plt.xlabel("count of objects")
plt.title("delete by one")
plt.legend()
plt.show()
```
